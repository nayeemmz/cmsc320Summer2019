```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_1samp
```

```python
# proportion of bot tweets given to us
p = 0.7
```

```python
# Sample 100 tweets
## This function chooses a value in  a vectore (0,1)
## with probability, p
## We need 100 samples from this vector with replacement,
## since there are fewer items in the vector than the size
## of the sample we are creating
x = np.random.choice([0,1],100,p=[1-p,p])
x
```

```python
## Estimated proportion that are bot generated
p_hat = np.mean(x)
p_hat
```

```python
# let us create a list with 1000 estimates
p_hats =  [np.mean(np.random.choice([0,1],100,p=[1-p,p])) for x in range(1000)]
p_hats
```

```python
len(p_hats)
```

```python
plt.hist(p_hats)
```

```python
plt.title("Distribution of p estimates from 100 tweets")
plt.xlabel("Estimate")
plt.ylabel("Frequency")
```
