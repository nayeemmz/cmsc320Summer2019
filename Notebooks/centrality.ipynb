```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
G.add_edges_from([('1','2'),('1','3'),('1','4'),('2','3'),('4','3'),('4','5'),('4','6'),('5','6'),\
                  ('5','7'),('5','8'),('6','7'),('6','8'),('7','8'),('7','9')])
nx.draw(G, with_labels=True)
plt.show()
```

```python
degree_centrality = nx.degree_centrality(G)
print(['%s %0.2f' %(node,degree_centrality[node]) for node in degree_centrality])
```

```python
closeness_centrality = nx.closeness_centrality(G)
print(['%s %0.2f' %(node,closeness_centrality[node]) for node in closeness_centrality])
```

```python
betweenness_centrality = nx.betweenness_centrality(G,normalized=False)
print(['%s %0.2f' %(node,betweenness_centrality[node]) for node in betweenness_centrality])
```

```python
G = nx.Graph()
G.add_edges_from([('A','B'),('A','C'),('A','D'),('A','E'),('E','F'),('F','H'),\
                  ('F','I'),('F','J'),('E','G'),('G','K'),('G','L'),('G','M')])
nx.draw(G, with_labels=True)
plt.show()
```

```python
centrality = nx.eigenvector_centrality(G)
print(['%s %0.2f' %(node,centrality[node]) for node in centrality])
```
